## Multiflow pattern : Via KERAS 

```python
from keras.models import Sequential
from keras.layers import Dense

import numpy
# fix random seed for reproducibility
numpy.random.seed(7)

dataset = numpy.loadtxt("BDOShoham.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:9]
Y = dataset[:,9]

# encode class values as integers
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
# create model
model = Sequential()
model.add(Dense(25, input_dim=9, kernel_initializer='normal', activation='relu'))
model.add(Dense(25, kernel_initializer='normal', activation='relu'))
model.add(Dense(25, kernel_initializer='normal', activation='relu'))
model.add(Dense(6, kernel_initializer='normal', activation='sigmoid'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model (takes several minutes)
model.fit(X, dummy_y, epochs=500, batch_size=10)

# evaluate the model
scores = model.evaluate(X, dummy_y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
